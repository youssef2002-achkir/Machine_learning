```python
t1 = time.time()
res = 0
for k in range(x.shape[0]):
    res = res + float(x[k] * y[k])
t2 = time.time()
print("took {} seconds".format(t2-t1))
print("Dot of x and y is :" , res)
```

```python
t1 = time.time()
res = np.dot(x.T,y)
t2 = time.time()
print("took {} seconds".format(t2-t1))
print("Dot shape: {}".format(res.shape))
print("Dot :{}".format(float(res)))
```

```python
A = np.random.rand(1999,1)
B = np.random.rand(1999,1)
assert A.shape == B.shape
print("A shape:", A.shape) 
print("B shape:", B.shape) 

```

```python
t1 = time.time()
res = 0
for i in range(A.shape[0]):
    res += float(A[i] * B[i]) 
t2 = time.time()
print("Naive method took us {} seconds".format(t2-t1))

t3 = time.time()
res = float(np.dot(A.T, B))
t4 = time.time()
print("Numpy method took us {} seconds".format(t4-t3))
```
